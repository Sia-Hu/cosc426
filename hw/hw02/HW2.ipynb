### Create Dataset

```python
import pandas as pd

sents = [{'sae': 'you seek the Force', 'yoda': 'the Force you seek'},
         {'sae': 'the Jedi follow the path', 'yoda': 'the path the Jedi follow'},
         {'sae': 'I teach you', 'yoda': 'you I teach'},
         {'sae': 'the Jedi fight', 'yoda': 'fight the Jedi do'},
         {'sae': 'you sleep', 'yoda': 'sleep you do'},
         {'sae': 'the Jedi fight in the night', 'yoda': 'in the night fight the Jedi do'},
         {'sae': 'the Jedi fight in the night in the rain', 'yoda': 'in the night in the rain fight the Jedi do'},
         {'sae': 'the Jedi fight in the night in the rain', 'yoda': 'in the rain in the night fight the Jedi do'},
         {'sae': 'the darkness in the night teach the Jedi', 'yoda': 'the Jedi the darkness in the night teach'},
         {'sae': 'the Jedi seek the Force through the darkness', 'yoda': 'the Force through the darkness the Jedi seek'}
        ]
    
def make_ungrammatical(sentence):
    words = sentence.split()
    if len(words) > 1:
        words[0], words[1] = words[1], words[0]
    
    ungrammatical = " ".join(words)
    if is_grammatical('hw2_grammar_sae.txt', ungrammatical):
        raise ValueError(f"Generated sentence is still grammatical: {ungrammatical}")
    return ungrammatical

rows = []
pair_id = 1
sent_id = 1

for sent in sents:
    sae_gram = sent['sae']
    sae_ungram = make_ungrammatical(sent['sae'])
    yoda = sent['yoda']
    
    rows.append({'sentid': sent_id, 'pairid': pair_id, 'comparison': 'expected', 'sentence': yoda})
    sent_id += 1
    rows.append({'sentid': sent_id, 'pairid': pair_id, 'comparison': 'unexpected', 'sentence': sae_gram})
    sent_id += 1
    rows.append({'sentid': sent_id, 'pairid': pair_id, 'comparison': 'sae_ungrammatical', 'sentence': sae_ungram})
    sent_id += 1
    
    pair_id += 1

dataset_df = pd.DataFrame(rows)

def save_dataset(dataset_df, filename):
    minimalPair_columns = ['sentid', 'pairid', 'comparison', 'sentence']
    
    output_df = dataset_df[minimalPair_columns].copy()
    output_df.to_csv(filename, sep='\t', index=False)
    
    print(f"Saved dataset to:") 
    return filename

save_dataset(dataset_df, "hw2_eval.tsv")
```

```python
import random

sae_grammar_file = 'hw2_grammar_sae.txt'
yoda_grammar_file = 'hw2_grammar_yoda.txt'

sae_pcfg = HW2.create_pcfg(sae_grammar_file)
yoda_pcfg = HW2.create_pcfg(yoda_grammar_file)

def create_ungrammatical_variations(sentence, grammar_file):
    words = sentence.split()
    variations = []
    
    if len(words) < 2:
        return variations
        
    # Swap adjacent words
    for i in range(min(3, len(words) - 1)):  
        modified = words.copy()
        modified[i], modified[i + 1] = modified[i + 1], modified[i]
        candidate = " ".join(modified)
        if not is_grammatical(grammar_file, candidate):
            variations.append(('adjacent_swap', candidate))
    
    # Swap first and last word
    if len(words) > 2:
        modified = words.copy()
        modified[0], modified[-1] = modified[-1], modified[0]
        candidate = " ".join(modified)
        if not is_grammatical(grammar_file, candidate):
            variations.append(('first_last_swap', candidate))
    
    # Random word order scrambling
    attempts = 0
    while attempts < 10:
        modified = words.copy()
        random.shuffle(modified)
        candidate = " ".join(modified)
        if candidate != sentence and not is_grammatical(grammar_file, candidate):
            variations.append(('scrambled', candidate))
            break
        attempts += 1
    
    return variations

controlled_pairs = controlled_pairs = [
    # Basic SVO patterns
    {'sae': 'the Jedi seek the Force', 'yoda': 'the Force the Jedi seek', 'phenomenon': 'OSV_basic'},
    {'sae': 'I follow the path', 'yoda': 'the path I follow', 'phenomenon': 'OSV_basic'},
    {'sae': 'they teach the Jedi', 'yoda': 'the Jedi they teach', 'phenomenon': 'OSV_basic'},
    {'sae': 'you seek the light', 'yoda': 'the light you seek', 'phenomenon': 'OSV_basic'},
    {'sae': 'we follow the darkness', 'yoda': 'the darkness we follow', 'phenomenon': 'OSV_basic'},
    
    # Intransitive verbs with auxiliary "do" 
    {'sae': 'the Jedi fight', 'yoda': 'fight the Jedi do', 'phenomenon': 'IV_with_aux'},
    {'sae': 'you sleep', 'yoda': 'sleep you do', 'phenomenon': 'IV_with_aux'},
    {'sae': 'they fight', 'yoda': 'fight they do', 'phenomenon': 'IV_with_aux'},
    {'sae': 'I sleep', 'yoda': 'sleep I do', 'phenomenon': 'IV_with_aux'},
    {'sae': 'we fight', 'yoda': 'fight we do', 'phenomenon': 'IV_with_aux'},
    
    # Prepositional phrases with transitive verbs
    {'sae': 'the Jedi seek the Force through the darkness', 'yoda': 'the Force through the darkness the Jedi seek', 'phenomenon': 'PP_TV_complex'},
    {'sae': 'you follow the path in the forest', 'yoda': 'the path in the forest you follow', 'phenomenon': 'PP_TV_complex'},
    {'sae': 'I teach the Jedi through the light', 'yoda': 'the Jedi through the light I teach', 'phenomenon': 'PP_TV_complex'},
    {'sae': 'they seek the path in the night', 'yoda': 'the path in the night they seek', 'phenomenon': 'PP_TV_complex'},
    
    # Prepositional phrases with intransitive verbs
    {'sae': 'the Jedi fight in the night', 'yoda': 'in the night fight the Jedi do', 'phenomenon': 'PP_IV_complex'},
    {'sae': 'you sleep through the rain', 'yoda': 'through the rain sleep you do', 'phenomenon': 'PP_IV_complex'},
    {'sae': 'they fight in the darkness', 'yoda': 'in the darkness fight they do', 'phenomenon': 'PP_IV_complex'},
    {'sae': 'we sleep in the forest', 'yoda': 'in the forest sleep we do', 'phenomenon': 'PP_IV_complex'},
    
    # NP with embedded PP
    {'sae': 'the Jedi in the forest seek the Force', 'yoda': 'the Force the Jedi in the forest seek', 'phenomenon': 'embedded_PP'},
    {'sae': 'the darkness in the night teach you', 'yoda': 'you the darkness in the night teach', 'phenomenon': 'embedded_PP'},
    {'sae': 'the path through the light follow I', 'yoda': 'the light through the path I follow', 'phenomenon': 'embedded_PP'},
    
    # Nested PP constructions
    {'sae': 'the Jedi seek the Force through the path in the forest', 'yoda': 'the Force through the path in the forest the Jedi seek', 'phenomenon': 'nested_PP_TV'},
    {'sae': 'you follow the light in the darkness through the night', 'yoda': 'the light in the darkness through the night you follow', 'phenomenon': 'nested_PP_TV'},
    {'sae': 'they teach the Jedi through the Force in the light', 'yoda': 'the Jedi through the Force in the light they teach', 'phenomenon': 'nested_PP_TV'},
    {'sae': 'I seek the path in the forest through the rain', 'yoda': 'the path in the forest through the rain I seek', 'phenomenon': 'nested_PP_TV'},
    {'sae': 'the Jedi fight in the forest through the night', 'yoda': 'in the forest through the night fight the Jedi do', 'phenomenon': 'nested_PP_IV'},
    {'sae': 'you sleep through the darkness in the rain', 'yoda': 'through the darkness in the rain sleep you do', 'phenomenon': 'nested_PP_IV'},
    {'sae': 'they fight in the night through the forest', 'yoda': 'in the night through the forest fight they do', 'phenomenon': 'nested_PP_IV'},
    {'sae': 'the Jedi in the forest seek the Force through the darkness in the night', 'yoda': 'the Force through the darkness in the night the Jedi in the forest seek', 'phenomenon': 'complex_nested'},
    {'sae': 'the darkness in the rain teach you through the path in the light', 'yoda': 'you through the path in the light the darkness in the rain teach', 'phenomenon': 'complex_nested'}
]

for pair in controlled_pairs:
    assert is_grammatical(sae_grammar_file, pair['sae']), f"SAE sentence not grammatical: {pair['sae']}"
    assert is_grammatical(yoda_grammar_file, pair['yoda']), f"Yoda sentence not grammatical: {pair['yoda']}"

# Create the dataset
rows = []
pair_id = 1
sent_id = 1
for pair in controlled_pairs:
    sae_gram = pair['sae']
    yoda = pair['yoda']
    phenomenon = pair['phenomenon']
    
    ungrammatical_variations = create_ungrammatical_variations(sae_gram, sae_grammar_file)
    
    rows.append({'sentid': sent_id, 'pairid': pair_id, 'comparison': 'expected', 'sentence': yoda, 'phenomenon': phenomenon})
    sent_id += 1
    rows.append({'sentid': sent_id, 'pairid': pair_id, 'comparison': 'unexpected', 'sentence': sae_gram, 'phenomenon': phenomenon})
    sent_id += 1
    for var_type, ungram in ungrammatical_variations:
        rows.append({'sentid': sent_id, 'pairid': pair_id, 'comparison': f'sae_ungrammatical_{var_type}', 'sentence': ungram, 'phenomenon': phenomenon})
        sent_id += 1
        
    pair_id += 1

dataset_df = pd.DataFrame(rows)
save_dataset(dataset_df, "hw2_systematic_eval.tsv")
```

```python
df_bypair = pd.read_csv('results/hw2_systematic_eval_bypair.tsv', sep='\t')

df_bypair['yoda_vs_grammatical'] = df_bypair['expected'] - df_bypair['unexpected']
df_bypair['abs_diff_grammatical'] = abs(df_bypair['yoda_vs_grammatical'])

print(f"Mean absolute difference (Yoda vs Grammatical): {df_bypair['abs_diff_grammatical'].mean():.4f}")
print(f"Median absolute difference: {df_bypair['abs_diff_grammatical'].median():.4f}")
print(f"Standard deviation: {df_bypair['abs_diff_grammatical'].std():.4f}")

ungrammatical_cols = [col for col in df_bypair.columns if 'sae_ungrammatical' in col]

# Store results for comparison
comparison_results = {}

for col in ungrammatical_cols:
    print(f"\n--- {col.replace('sae_ungrammatical_', '').upper()} ---")
    
    valid_mask = df_bypair[col].notna()
    if valid_mask.sum() == 0:
        continue
        
    valid_data = df_bypair[valid_mask]
    
    # Calculate differences and absolute differences
    diff_values = valid_data['expected'] - valid_data[col]
    abs_diff_values = abs(diff_values)
    
    print(f"Mean absolute difference (Yoda vs {col}): {abs_diff_values.mean():.4f}")
    print(f"Median absolute difference: {abs_diff_values.median():.4f}")
    print(f"Standard deviation: {abs_diff_values.std():.4f}")
    
    # For each valid pair, determine which is closer
    gram_abs_diff = abs(valid_data['yoda_vs_grammatical'])
    ungram_abs_diff = abs_diff_values
    
    closer_to_grammatical = (gram_abs_diff < ungram_abs_diff).sum()
    closer_to_ungrammatical = (ungram_abs_diff < gram_abs_diff).sum()
    tied = (gram_abs_diff == ungram_abs_diff).sum()
    total_valid = len(valid_data)
    
    print(f"Pairs closer to grammatical: {closer_to_grammatical} ({closer_to_grammatical/total_valid:.1%})")
    print(f"Pairs closer to {col}: {closer_to_ungrammatical} ({closer_to_ungrammatical/total_valid:.1%})")
    print(f"Tied: {tied}")
    
    # Store for overall comparison
    comparison_results[col] = {
        'closer_to_grammatical': closer_to_grammatical,
        'closer_to_ungrammatical': closer_to_ungrammatical,
        'total_valid': total_valid,
        'mean_abs_diff': abs_diff_values.mean()
    }

# Overall interpretation across all ungrammatical types
if comparison_results:
    total_closer_to_gram = sum([result['closer_to_grammatical'] for result in comparison_results.values()])
    total_closer_to_ungram = sum([result['closer_to_ungrammatical'] for result in comparison_results.values()])
    total_comparisons = total_closer_to_gram + total_closer_to_ungram
    
    print(f"\n=== SUMMARY ===")
    print(f"Total pair comparisons across all ungrammatical types: {total_comparisons}")
    print(f"Pairs closer to grammatical: {total_closer_to_gram} ({total_closer_to_gram/total_comparisons:.1%})")
    print(f"Pairs closer to ungrammatical: {total_closer_to_ungram} ({total_closer_to_ungram/total_comparisons:.1%})")
    
    if total_closer_to_gram > total_closer_to_ungram:
        interpretation = "DistilGPT2 treats Yoda-speak as closer to grammatical English than to ungrammatical English"
    else:
        interpretation = "DistilGPT2 treats Yoda-speak as closer to ungrammatical English than to grammatical English"
else:
    interpretation = "Cannot determine grammaticality status due to insufficient ungrammatical data"

print(f"\nOverall Interpretation: {interpretation}.")
```

```python
df_bypair = pd.read_csv('results/hw2_systematic_eval_4_bypair.tsv', sep='\t')

df_bypair['yoda_vs_grammatical'] = df_bypair['expected'] - df_bypair['unexpected']
df_bypair['abs_diff_grammatical'] = abs(df_bypair['yoda_vs_grammatical'])

print(f"Mean absolute difference (Yoda vs Grammatical): {df_bypair['abs_diff_grammatical'].mean():.4f}")
print(f"Median absolute difference: {df_bypair['abs_diff_grammatical'].median():.4f}")
print(f"Standard deviation: {df_bypair['abs_diff_grammatical'].std():.4f}")

ungrammatical_cols = [col for col in df_bypair.columns if 'sae_ungrammatical' in col]

# Store results for comparison
comparison_results = {}

for col in ungrammatical_cols:
    print(f"\n--- {col.replace('sae_ungrammatical_', '').upper()} ---")
    
    valid_mask = df_bypair[col].notna()
    if valid_mask.sum() == 0:
        continue
        
    valid_data = df_bypair[valid_mask]
    
    # Calculate differences and absolute differences
    diff_values = valid_data['expected'] - valid_data[col]
    abs_diff_values = abs(diff_values)
    
    print(f"Mean absolute difference (Yoda vs {col}): {abs_diff_values.mean():.4f}")
    print(f"Median absolute difference: {abs_diff_values.median():.4f}")
    print(f"Standard deviation: {abs_diff_values.std():.4f}")
    
    # For each valid pair, determine which is closer
    gram_abs_diff = abs(valid_data['yoda_vs_grammatical'])
    ungram_abs_diff = abs_diff_values
    
    closer_to_grammatical = (gram_abs_diff < ungram_abs_diff).sum()
    closer_to_ungrammatical = (ungram_abs_diff < gram_abs_diff).sum()
    tied = (gram_abs_diff == ungram_abs_diff).sum()
    total_valid = len(valid_data)
    
    print(f"Pairs closer to grammatical: {closer_to_grammatical} ({closer_to_grammatical/total_valid:.1%})")
    print(f"Pairs closer to {col}: {closer_to_ungrammatical} ({closer_to_ungrammatical/total_valid:.1%})")
    print(f"Tied: {tied}")
    
    # Store for overall comparison
    comparison_results[col] = {
        'closer_to_grammatical': closer_to_grammatical,
        'closer_to_ungrammatical': closer_to_ungrammatical,
        'total_valid': total_valid,
        'mean_abs_diff': abs_diff_values.mean()
    }

if comparison_results:
    total_closer_to_gram = sum([result['closer_to_grammatical'] for result in comparison_results.values()])
    total_closer_to_ungram = sum([result['closer_to_ungrammatical'] for result in comparison_results.values()])
    total_comparisons = total_closer_to_gram + total_closer_to_ungram
    
    print(f"\n=== SUMMARY ===")
    print(f"Total pair comparisons across all ungrammatical types: {total_comparisons}")
    print(f"Pairs closer to grammatical: {total_closer_to_gram} ({total_closer_to_gram/total_comparisons:.1%})")
    print(f"Pairs closer to ungrammatical: {total_closer_to_ungram} ({total_closer_to_ungram/total_comparisons:.1%})")
    
    if total_closer_to_gram > total_closer_to_ungram:
        interpretation = "DistilGPT2 treats Yoda-speak as closer to grammatical English than to ungrammatical English"
    else:
        interpretation = "DistilGPT2 treats Yoda-speak as closer to ungrammatical English than to grammatical English"
else:
    interpretation = "Cannot determine grammaticality status due to insufficient ungrammatical data"

print(f"\nOverall Interpretation: {interpretation}.")
```
