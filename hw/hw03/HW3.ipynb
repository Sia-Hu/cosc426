When we try different smoothing methods, we introduce hyperparameters. If we only evaluate on the training set, the model will overfit.  

```python
tokens = training_data

split_idx = int(0.9 * len(tokens))
train_tokens = tokens[:split_idx]
val_tokens = tokens[split_idx:]

print(f"Train size: {len(train_tokens)}, Validation size: {len(val_tokens)}")

unigram_freqs = HW3.get_ngramFreqs(train_tokens, 1)
bigram_freqs = HW3.get_ngramFreqs(train_tokens, 2)
trigram_freqs = HW3.get_ngramFreqs(train_tokens, 3)

total_tokens = len(train_tokens)
vocab_size = len(HW3.get_hf_vocab("distilgpt2"))
```

```python
import math

def get_interpolated_prob(word, context, lambdas, unigram_freqs, bigram_freqs, trigram_freqs, total_tokens, vocab_size):
    
    l1, l2, l3 = lambdas
    
    unigram_prob = unigram_freqs.get((word,), 0) / total_tokens
    
    if len(context) >= 1:
        prev_word = (context[-1],)
        count_bigram = bigram_freqs.get((context[-1], word), 0)
        count_prev = unigram_freqs.get(prev_word, 0)
        bigram_prob = (count_bigram / count_prev) if count_prev > 0 else 0.0
    else:
        bigram_prob = 0.0
    
    if len(context) >= 2:
        prev_bigram = (context[-2], context[-1])
        count_trigram = trigram_freqs.get((context[-2], context[-1], word), 0)
        count_prev2 = bigram_freqs.get(prev_bigram, 0)
        trigram_prob = (count_trigram / count_prev2) if count_prev2 > 0 else 0.0
    else:
        trigram_prob = 0.0
    
    return l1*unigram_prob + l2*bigram_prob + l3*trigram_prob

def evaluate_interpolation(text, lambdas, unigram_freqs, bigram_freqs, trigram_freqs, total_tokens, vocab_size):
    
    log_prob_sum = 0.0
    count = 0
    for i in range(2, len(text)):
        context = text[i-2:i]
        word = text[i]
        prob = get_interpolated_prob(word, context, lambdas, unigram_freqs, bigram_freqs, trigram_freqs, total_tokens, vocab_size)
        if prob > 0:
            log_prob_sum += math.log(prob)
        else:
            log_prob_sum += math.log(1e-10)
        count += 1
    avg_log_prob = log_prob_sum / count
    perplexity = math.exp(-avg_log_prob)
    return perplexity
```

```python
smoothing_methods = ["MLE", "add-0.001", "add-0.01", "add-0.1", "add-1"]
lambda_candidates = [
    (0.1, 0.3, 0.6),
    (0.2, 0.3, 0.5),
    (0.3, 0.3, 0.4),
    (0.33, 0.33, 0.34)
]

results = []

# Bigram
for smooth in smoothing_methods:
    pp = HW3.evaluate(val_tokens, smooth, 2, bigram_freqs, unigram_freqs, vocab_size)
    results.append(("bigram", smooth, pp))

# Trigram
for smooth in smoothing_methods:
    pp = HW3.evaluate(val_tokens, smooth, 3, trigram_freqs, bigram_freqs, vocab_size)
    results.append(("trigram", smooth, pp))

# Interpolation
for lambdas in lambda_candidates:
    pp = evaluate_interpolation(val_tokens, lambdas, unigram_freqs, bigram_freqs, trigram_freqs, total_tokens, vocab_size)
    results.append(("interpolation", lambdas, pp))

results_sorted = sorted(results, key=lambda x: x[2])

print("Validation Results")
for model, smooth, pp in results_sorted:
    if model == "interpolation":
        print(f"{model}, λ={smooth}: Perplexity={pp:.2f}")
    else:
        print(f"{model}, {smooth}: Perplexity={pp:.2f}")

best_model = results_sorted[0]
print("\nBest model on validation set:", best_model)
```

```python
# Evaluate best model on other files
test_results = []
for testfile_name, tokenized_testfile in [("through_the_looking_glass.txt", looking_glass_data), ("sherlock_holmes_short.txt", sherlock_holmes_data)]:
    pp = HW3.evaluate(tokenized_testfile, 'add-0.01', 2, bigram_freqs, unigram_freqs, vocab_size)
    test_results.append((testfile_name, pp))
    
print("Test Results on Best Model (Bigram with add-0.01)")
for testfile_name, pp in test_results:
    print(f"{testfile_name}: Perplexity={pp:.2f}")
```

```python
def txt_to_tsv(txt_path, tsv_path):
    with open(txt_path, "r", encoding="utf-8") as f:
        lines = [line.strip() for line in f if line.strip()]
    
    text = " ".join(lines)
    
    df = pd.DataFrame({
        "sentid": [0],       
        "sentence": [text]
    })
    
    df.to_csv(tsv_path, sep="\t", index=False, encoding="utf-8")

txt_to_tsv("data/alice_in_wonderland.txt", "data/alice_eval.tsv")
txt_to_tsv("data/through_the_looking_glass.txt", "data/lookingglass_eval.tsv")
txt_to_tsv("data/sherlock_holmes_short.txt", "data/sherlock_eval.tsv")
```

```python
import math

def evaluate_lm_tsv(tsv_path: str):
    """
    Evaluate perplexity from a prediction TSV file.
    Assumes the file has a column 'prob' with P(word|context).
    """
    df = pd.read_csv(tsv_path, sep="\t")
    
    probs = df["prob"].astype(float)
    probs = probs[probs > 0]   
    
    log_prob_sum = sum(math.log(p) for p in probs)
    avg_log_prob = log_prob_sum / len(probs)
    perplexity = math.exp(-avg_log_prob)
    
    return perplexity
```

Our bigram and trigram models need to be directly comparable to the distilgpt2-based models. Different tokenizers produce different token sequences, which affects both the context and the counts in n-gram models.

```python
import nltk
nltk.download('punkt')
from nltk import word_tokenize

# use nltk to tokenize data/alice_in_wonderland.txt
with open("data/alice_in_wonderland.txt", "r", encoding="utf-8") as f:
    text = f.read()
nltk_tokens = word_tokenize(text)
print(nltk_tokens[:20])
```

NLTK tokenizers split text mostly at whitespace and punctuation. From Distilgpt2 uses a subword tokenizer,Byte-Pair Encoding (https://huggingface.co/distilbert/distilgpt2). For example 'Alice' is separated into 'Ġal' and 'ice' here in the training data.

Bigram and trigram models rely on sequences of tokens. If the tokenization differs, the exact n-grams counted will differ.

Perplexity depends on probabilities of token sequences. Using different tokenizations would make perplexities incomparable, because the models are effectively predicting probabilities over different sets of tokens.

```python
# Tokenizers
def nltk_tokenize(text):
    return word_tokenize(text)

def char_tokenize(text):
    return list(text)

# Vocabularies
def get_glove_vocab(filepath="data/glove_vocab.txt"):
    with open(filepath, 'r', encoding='utf-8') as f:
        vocab = set(line.strip() for line in f if line.strip())
    return vocab

def get_training_vocab(tokens):
    return set(tokens)

# Preprocessing
def preprocess_with_tokenizer(filepath, tokenizer_func):
    with open(filepath, 'r', encoding='utf-8') as f:
        text = f.read()
    tokens = tokenizer_func(text)
    return tokens

# Data preparation
alice_nltk = preprocess_with_tokenizer("data/alice_in_wonderland.txt", nltk_tokenize)
alice_char = preprocess_with_tokenizer("data/alice_in_wonderland.txt", char_tokenize)

looking_glass_nltk = preprocess_with_tokenizer("data/through_the_looking_glass.txt", nltk_tokenize)
sherlock_nltk = preprocess_with_tokenizer("data/sherlock_holmes_short.txt", nltk_tokenize)

looking_glass_char = preprocess_with_tokenizer("data/through_the_looking_glass.txt", char_tokenize)
sherlock_char = preprocess_with_tokenizer("data/sherlock_holmes_short.txt", char_tokenize)

glove_vocab = get_glove_vocab("data/glove_vocab.txt")
nltk_training_vocab = get_training_vocab(alice_nltk)
char_training_vocab = get_training_vocab(alice_char)

# Training and evaluation
results = []

configurations = [
    ("NLTK", alice_nltk, 
     {"looking_glass": looking_glass_nltk, "sherlock": sherlock_nltk},
     "GloVe", len(glove_vocab)),
    ("NLTK", alice_nltk, 
     {"looking_glass": looking_glass_nltk, "sherlock": sherlock_nltk},
     "Training", len(nltk_training_vocab)),
    
    ("Character", alice_char, 
     {"looking_glass": looking_glass_char, "sherlock": sherlock_char},
     "Training", len(char_training_vocab)),
]


for tokenizer_name, train_tokens, test_files, vocab_name, vocab_size in configurations:    
    for n in [2, 3]:
        ngram_freqs = HW3.get_ngramFreqs(train_tokens, n)
        n_minus1_freqs = HW3.get_ngramFreqs(train_tokens, n-1)
        
        for test_name, test_tokens in test_files.items():
            ppl = HW3.evaluate(test_tokens, "add-0.001", n, ngram_freqs, n_minus1_freqs, vocab_size)
            
            results.append({
                "Tokenizer": tokenizer_name,
                "Vocabulary": vocab_name,
                "N-gram": f"{n}-gram",
                "Test File": test_name,
                "Perplexity": ppl
            })

df_results = pd.DataFrame(results)

summary = df_results.pivot_table(
    index=["Tokenizer", "Vocabulary", "Test File"],
    columns="N-gram",
    values="Perplexity"
)
print(summary.round(2))
```

The results reveal differences in how tokenization and vocabulary choices affect n-gram model performance. 

Character-level tokenization eliminates out-of-vocabulary issues and captures morphological patterns that generalize well across domains. In contrast, NLTK tokenization with GloVe vocabulary produces catastrophically high perplexities, suggesting massive vocabulary mismatch where many domain-specific words from the old literature are absent from the modern general-purpose GloVe vocabulary. When NLTK uses training vocabulary instead, performance improves, though still far worse than character-level models.

Word-level trigrams consistently worsen perplexity. There aren't enough training examples to reliably estimate word trigram probabilities.
